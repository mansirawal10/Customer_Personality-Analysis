STEP 2: Import Libraries


```python
import pandas as pd
import numpy as np
```

STEP 3: Load the Dataset

```python
df = pd.read_csv("marketing_campaign.csv", sep="\t")
df.head()
```

STEP 4: Basic Info


```python
df.shape
df.info()
df.describe()
```

STEP 5: Rename Columns for Consistency

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

STEP 6: Check & Handle Missing Values

```python
df.isnull().sum()
```

```python
df['income'] = df['income'].fillna(df['income'].median())
```

STEP 7: Remove Duplicates

```python
df = df.drop_duplicates()
```

STEP 8: Convert Date Columns

```python
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format='%d-%m-%Y')
```

STEP 9: Standardize Text Columns

```python
df['marital_status'] = df['marital_status'].str.title().str.strip()
df['education'] = df['education'].str.title().str.strip()
```

STEP 10: Fix Data Types

```python
df['income'] = df['income'].astype(float)
df['year_birth'] = df['year_birth'].astype(int)
```

STEP 12: Save Cleaned Dataset

```python
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```
